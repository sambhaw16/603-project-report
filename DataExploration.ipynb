## Let's see if we can learn a decision tree that predicts Weapon

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
df2 = df.copy()
```

```python
df2.head(10)
```

```python
df2.drop(['CrimeCode', 'Description', 'Post'], axis=1, inplace=True)
```

```python
df2.head(10)
```

```python
len(df2)
```

```python
df2.dropna(inplace=True)
```

```python
len(df2)
```

```python
df2.head(10)
```

## The scikit-learn decison tree wants EVERYTHING to be numbers, which is a pain.  So we need to convert somehow.  We'll use a one-hot encoding.

```python
pd.get_dummies(df2['Inside/Outside'], prefix = 'Inside_Outside')
```

## OK, that looks good.  Now lets build up a dataframe for the X (predictor) variables, and the Y (predicted) variable.

```python
X = pd.get_dummies(df2['Inside/Outside'], prefix = 'Inside_Outside')
```

```python
X = pd.concat([X, pd.get_dummies(df2['District'], prefix = 'District')], axis=1)
```

```python
X = pd.concat([X, pd.get_dummies(df2['CrimeDay'], prefix = 'CrimeDay')], axis=1)
```

```python
X.head(3)
```

## What about CrimeHour?  Is it already a number?

```python
df2.dtypes
```

## Nope.  Let's convert it and add it to our dataframe.

```python
X['CrimeHour'] = df2['CrimeHour'].astype(float)
```

## What do things look like now?

```python
X.head(10)
```

## Nice!  Now let's get a Y dataframe based on the type of weapon.

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
Y = le.fit_transform(df2['Weapon'])
```

```python
le.inverse_transform([0, 1, 2, 3])
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X.values, Y)
```

```python
X.columns
```

```python
with open("dt.dot", 'w') as f:
    export_graphviz(dt, out_file=f, feature_names=X.columns)
```

## Too big! Need to depth limit the tree.

```python
dt = DecisionTreeClassifier(max_depth = 4)
dt.fit(X.values, Y)
with open("dt.dot", 'w') as f:
    export_graphviz(dt, out_file=f, feature_names=X.columns)
os.system('dot -Tpng dt.dot -o dt.png')
Image(filename='dt.png')
```

## How good is the tree?  Let's try cross validation.

```python
from sklearn import cross_validation
```

```python
scores = cross_validation.cross_val_score(dt, X, Y, cv = 10)
```

```python
scores
```

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

## Again, the question is "how good is that"?  What would you get by random guessing?

```python
pd.value_counts(Y)
```

```python
46415. / len(Y)
```

## The answer is that we're doing no better than random guessing!  Maybe try a deeper tree.

```python
dt = DecisionTreeClassifier(min_samples_split = 100)
scores = cross_validation.cross_val_score(dt, X, Y, cv = 10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
